This file contains the code for exploring and cleaning the dataset, including the tasks:
- checking for missing values.
- exploring data distributions. 
- removing unnecessary columns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

**Load the dataset**

```python
# Get the current working directory & Construct the relative path to train.csv
current_dir = os.getcwd() 
relative_path = os.path.join('..', 'data', 'train.csv')
```

```python
training_data = pd.read_csv(os.path.join(current_dir, relative_path))
```

**Explore the dataset**

```python
print(training_data.columns.size)
print(training_data.columns)
```

```python
for column in training_data.columns:
    print(f"Unique values in '{column}': {training_data[column].unique()}\n")
```

```python
print(training_data.isnull().sum())

# The Arrival Delay in Minutes column has 310 missing values.
```

**Remove rows with null values**

```python
training_data.dropna(inplace=True)
print(training_data.isnull().sum())
```

**Remove unnecessary columns** 

```python
# Drop the 'Unnamed: 0' and 'id' columns
training_data = training_data.drop(columns=['Unnamed: 0', 'id'])
```

**Save cleaned data**

```python
relative_path = os.path.join('..', 'data', 'cleaned_train_data.csv')
training_data.to_csv(os.path.join(current_dir, relative_path), index=False) # exclude the DataFrame index from being saved to the CSV file.
```
